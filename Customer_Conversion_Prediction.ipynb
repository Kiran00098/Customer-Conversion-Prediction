## Correction for outliers in dur column

```python
df.dur.describe()
```

```python
#IQR
q3 = df.dur.quantile(0.75)
q1 = df.dur.quantile(0.25)
iqr = q3 - q1 
```

```python
#Thresholds
q3 + (1.5 * iqr)
```

```python
q1 - (1.5*iqr)
```

```python
df.dur = df.dur.clip(0,643) 
```

```python
x = df[['age',"job",'marital', 'education_qual', 'call_type', 'day', 'mon',
       'dur', 'num_calls',"prev_outcome"]].values

y = df["y"].values
```

```python
from imblearn.combine import SMOTEENN
smt = SMOTEENN(sampling_strategy = "all")
x_smt,y_smt = smt.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_smt,y_smt,test_size=0.2)
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(x_train)
x_train = scalar.transform(x_train)
x_test = scalar.transform(x_test)
```
